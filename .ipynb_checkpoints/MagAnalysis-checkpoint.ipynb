```python
#Import packages
import numpy as np
import matplotlib.pyplot as plt
import csv
import pandas as pd
import seaborn as sns
```

```python
# Read data
sdata = pd.read_csv(r"D:\BSC IT DOCX\Research Papers\Research Internship-Columbia\Indoor localization\New Dataset\magneticnewcoordinate00.csv")
```

```python
# Insert column names
sdata.columns = ['TimeStamp', 'X', 'Y', 'Z', 'Magnitude']
#sdata.read_csv("D:\BSC IT DOCX\Research Papers\Research Internship-Columbia\Indoor localization\magneticindoorlocalizationdata.csv", index=False)
```

```python
#Convert the timestamp to its correct data type
sdata['TimeStamp'] =  pd.to_datetime(sdata['TimeStamp'], format='%Y-%m-%d %H:%M:%S:%f')
```

```python
sdata.info()
```

#### Visualizing all the readings

```python
time = sdata['TimeStamp']
ax = sdata['X']
ay = sdata['Y']
az = sdata['Z']
amag = sdata['Magnitude']
```

```python
plt.figure(figsize = (15, 5))
plt.plot(time, ax, time, ay, time, az, time, az, time, amag)
plt.title('Magnetometer data', fontsize = 20)
plt.xlabel('Time', fontsize = 20)
plt.ylabel('Mag. Values', fontsize = 20)
```

```python
plt.figure(figsize = (15, 5))

plt.subplot(2, 2, 1)
plt.plot(ax, color = 'r')
plt.title('a[x]', fontsize = 10)


plt.subplot(2, 2, 2)
plt.plot(ay, color = 'g')
plt.title('a[y]', fontsize = 10)


plt.subplot(2, 2, 3)
plt.plot(ax, color = 'b')
plt.title('a[z]', fontsize = 10)


plt.subplot(2, 2, 4)
plt.plot(amag, color = 'k')
plt.title('a[mag]', fontsize = 10)
```

```python
x_mean = sdata[['X']].mean()
y_mean = sdata[['Y']].mean()
z_mean = sdata[['Z']].mean()
mag_mean = sdata[['Magnitude']].mean()
print(x_mean)
print(y_mean)
print(z_mean)
print(mag_mean)
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(6,5))
left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
ax = fig.add_axes([left, bottom, width, height]) 

start, stop = 0, 5

x_vals = np.linspace(start, stop)
y_vals = np.linspace(start, stop)
X, Y = np.meshgrid(x_vals, y_vals)
print(X)
print(X.shape)
x_mean = -19.44287213
Z=np.full_like(X, x_mean)
print(Z)
print(Z.shape)

cp = plt.contourf(X, Y, Z, levels=np.linspace(-100, 0, 80))
plt.colorbar(cp)

ax.set_title('Contour Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

```python
v = np.ones((50,50))
v[4][2] = mag_mean
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(6,5))
left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
ax = fig.add_axes([left, bottom, width, height]) 

start, stop = 0, 5

x_vals = np.linspace(start, stop)
y_vals = np.linspace(start, stop)
X, Y = np.meshgrid(x_vals, y_vals)
print(X)
print(X.shape)

Z= v

cp = plt.contourf(X, Y, Z, levels=np.linspace(-10, 90, 80))
plt.colorbar(cp)

contours = plt.contour(X, Y, Z, levels=[x_mean], color='black')
plt.clabel(contours, inline = True, fontsize = 8)
plt.imshow(Z, extent = [0, 5, 0, 5], origin = 'lower', cmap = plt.cm.jet, alpha = 0.5)

ax.set_title('Contour Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(6,5))
left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
ax = fig.add_axes([left, bottom, width, height]) 

start, stop, nvals = 0, 5, 2

x_vals = np.linspace(start, stop, nvals)
y_vals = np.linspace(start, stop, nvals)
X, Y = np.meshgrid(x_vals, y_vals)
print(X)
print(X.shape)
Z = np.array((x_mean, y_mean, z_mean, mag_mean)).reshape((2,2))
print(Z.shape)

cp = plt.contourf(X, Y, Z, 20, cmap = plt.cm.jet)
plt.colorbar(cp)
contours = plt.contour(X, Y, Z, 15, colors = 'black')
plt.clabel(contours, fontsize = 8)
#plt.imshow(Z, extent = [0, 5, 0, 5], origin = 'left', cmap = plt.cm.jet, alpha = 0.5)
plt.axis(aspect = 'image')
plt.show()
```
